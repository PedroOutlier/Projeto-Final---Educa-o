# Análise de Dados do ENEM

Este notebook faz uma análise simples sobre os dados do ENEM.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leitura dos dados
df = pd.read_csv('/content/enem_simulado.csv')
df.head()
```

```python
# Faltas por estado
faltas_estado = df.groupby('UF')[['FALTANTE_DIA1', 'FALTANTE_DIA2']].sum()
faltas_estado.plot(kind='bar', title='Faltas por Estado no ENEM')
plt.ylabel('Quantidade de Faltas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Médias de nota por estado
notas_estado = df.groupby('UF')[['NOTA_MT', 'NOTA_CN', 'NOTA_CH', 'NOTA_LC', 'NOTA_REDACAO']].mean()
notas_estado.plot(kind='bar', title='Notas Médias por Estado')
plt.ylabel('Nota Média')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
